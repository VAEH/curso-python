#Group BY

En esta clase vamos a aprender a como agrupar nuestras categorías y sobre esta aplicar algunas funciones estadísticas e inclusive vamos a aprender que con aggregate podemos crear nuestra propia función y aplicarla.

```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
pd.options.display.float_format = '{:,.3f}'.format
```

```python
#Libreria de visualización 
import seaborn as sns
```

```python
df = sns.load_dataset('diamonds')
df
```

Se agrupan los datos dado una variable, sobre este se pueden aplicar variables como el promedio. 

```python
df.groupby('cut').mean()
```

Se puede calcular otros estimadores estadisticos. 

```python
df.groupby('cut').median()
```

Se puede aplicar ese tipo de funciones a nivel de columnas, simplemente colocando la columna después del `Group By`. Se puede calcular la media, mediana, contar los datos según su grupo, etc. 

```python
df.groupby('cut')['carat'].count()
```

```python
df.groupby('cut')['carat'].max()
```

```python
df.groupby('cut')['carat'].min()
```

Se puede entender el `GroupBy` como un objeto que contiene paquetes y sobre estos paquetes se recorren aplicando diferentes funciones. 

```python
for i in [0,1,2,3]:
  print(i)
```

Se pueden aplicar funciones a nivel de Bucle

```python
for key_group, group in df.groupby('cut'):
  #De cada uno de esos grupos, seleccionare el promedio. 
  grouped_price = group['price'].mean()

  print('Cut: {}, Price {}\n'.format(key_group, grouped_price))
```

Se puede agrupar usando varios parametros. En este caso se realiza una agrupación por corte y color. 

Obtenemos un objeto que contiene dos indices, que contiene la variable de corte y la variable de color 

lo que hacemos con to_frame es convertir una serie de datos a tipo DataFrame, es decir a un marco de datos.
.
El mismo resultado lo podemos obtener al poner corchete doble [[ ]]
.
Ej: [ [ ‘price’ ] ] #nos prsenta la información como DataFrame
[‘price’] #nos presenta la información como serie de datos. 

```python
df.groupby(['cut', 'color'])['price'].mean().to_frame()
```

Se pueden agregar nuestras propias funciones o pedirle que se agreguen varias funciones de una sola vez.

Con `.aggregate()` se define una lista de las funciones que le quiero aplicar, en el agrupamiento se convierten en columnas. 

```python
df.groupby(['cut', 'color'])['price'].aggregate(['min', np.mean, max])
```

Se hace una función en promedio de unidades de Kilo

```python
def mean_kilo(x):
  return np.mean(x)/1000 
```

la función que se creo, se aplica a cada uno de estos grupos

```python
df.groupby(['cut', 'color'])['price'].aggregate(['min', np.mean, max, mean_kilo]).head(10)
```

Se puede crear un diccionario de instrucciones que podemos añadir al `aggregate`.

Se quiere crear varios calculos estadisticos para diferentes categorias. 

```python
dict_agg = {'carat':[min,max], 'price':[np.mean, mean_kilo]}
dict_agg
```

```python
df.groupby(['cut', 'color']).aggregate(dict_agg)
```

### Creación de Filtros

Función de filtro, se coge la función que ya se había creado anteriormente, se pide que cada vez que calcule en el promedio y lo divida entre 1000 del precio, seleccione solo aquellos datos mayores a 4. RETORNA BOOL

```python
def f_filter(x):
  return mean_kilo(x['price'])>4
```

Se utiliza `filter` y se entrega la función de filtro que se creo anteriormente.

Se obtiene que por cada grupo de categoria, se va a obtener aquellos elementos que satisfacen mi condición previamente definida, es decir, cuyo precio promedio de grupo dividio entre 1000 sea mayor a 4. 

```python
#X: CADA GRUPO POR TIPO DE CUT: 5 GRUPOS: IDEAL, PREMIUM, VERY GOOD, GOOD
print("df.groupby('cut').filter(f_filter)")
df.groupby('cut').filter(f_filter) 
```

```python
df.groupby('cut').filter(f_filter)['cut'].unique() 
```
