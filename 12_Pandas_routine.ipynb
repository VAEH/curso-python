````
import pandas as pd
import numpy as np
````

````
df_temp = pd.read_csv('GlobalLandTemperaturesByCountry.csv')
df_temp
````

Verificamos el tamaño del dataframe 

````
df_temp.shape
````

Vemos como esta compuesta nuestra base de datos

````
df_temp.info()
````

`unique()` vemos los paises que componen nuestra set de datos

````
df_temp['Country'].unique()
````

Vamos a mejorar nuestro formato base, convertir la columna `dt`  a una de tipo tiempo 

````
df_temp['year'] = pd.to_datetime(df_temp['dt'])
df_temp['year']
````

Creamos un filtro, donde seleccionamos las fechas mayores a 1970. Al realizar el valor, nos genera un indice, el cual vamos a utilizar para realizar el filtro. 

````
idx = df_temp['year'] > pd.to_datetime('1970-01-01')
df_temp = df_temp[idx]
df_temp
````

Utilizamos `groupby` para agrupar por fecha y por país y encontrar un promedio anual de temperatura. Utilizamos `Grouper` para agrupar por fecha. Elegimos por año y una frecuencia de cada año. con `agg` agregaamos los valores estadisticos que queremos aplicar a la columna **AverageTemperature**

````
df_t_avg = df_temp.groupby(['Country',
                 pd.Grouper(key='year', freq='1Y')                 
]).agg({'AverageTemperature':[np.mean,np.median]})
df_t_avg
````

utilizamos`.xsx` y esto nos permite seleccionar algún  país y seleccionamos la columna principal *AverageTmperature*

````
df_t_avg.xs('Colombia')['AverageTemperature'].plot()
````

Construir un dataset más compacto 

````
df_t_med = df_t_avg['AverageTemperature'][['median']].reset_index()
df_t_med
````

Para extraer el año de una fecha, usamos `dt.year`

````
df_t_med['date'] = df_t_med['year'].dt.year
df_t_med['date']
````

Se renombran algunas columnas y utilizamos `inplace=True` para no tener que escribir el data Frame 

````
df_t_med.rename(columns={'median':'temperature'}, inplace=True)
df_t_med
````

Para explorar la composición de las columnas por país, podemos hacer un BoxPlot que es una herramienta muy útil para encontrar la distribucción de las variables. Hacemos una tabla Pivot y lo hacemos con la tabla `pivot_table`. Los valores van hacer las temperaturas, el indice van  hacer las fechas y las columnas van hacer los paises. 

````
df_t_pivot = df_t_med.pivot_table(values='temperature', index='date', columns='Country')
df_t_pivot
````

Hacemos el Boxplot

````
df_t_pivot.boxplot()
````

Seleccionamos un número de paises más pequeño `sample` selecciona filas aleatorias de un dataframe. Utilizamos la traspuesta `T`

````
df_t_pivot.T.sample(10).T.boxplot(figsize=(20,15),rot=45)
````

Usamos la función `loc`, cogemos el df de agricultura. Usamos la función `melt` para volver las columnas partes del dataframe como una serie y escribimos las variables que queremos continuar conservadrvando como serie. El `melt` lo que hace es que antes todo lo que era varias columnas lo agrega a una sola, en este caso agrega los demás años a una sola ccolumna. 

````
df_agri.loc[:,cols].melt(id_vars=['Country Name','Country Code']).rename(
    columns={'variable':'date',
             'Country Name':'Country',
             'Country Code':'name',
             'value':'agriculture'})
````

Como se debe cambiar el formato a los demás df's, entonces se contruye una función que realice todo eso y nos ahorre el trabajo. 

````
def fun_format(df, col = 'agriculture'):
  return df.loc[:,cols].melt(id_vars=['Country Name','Country Code']).rename(
      columns={'variable':'date',
              'Country Name':'Country',
              'Country Code':'name',
              'value':col})
````

````
df_agri = fun_format(df_agri, col = 'agriculture')
df_fore = fun_format(df_fore, col = 'forest')
df_elec = fun_format(df_elec, col = 'electricprod')
df_co2e = fun_format(df_co2e, col = 'co2')
df_popu = fun_format(df_popu, col = 'population')
df_popu
````

````
df_t_med
````

````
df_popu['date'] = df_popu['date'].astype(float)
df_fore['date'] = df_fore['date'].astype(float)
df_agri['date'] = df_agri['date'].astype(float)
df_elec['date'] = df_elec['date'].astype(float)
df_co2e['date'] = df_co2e['date'].astype(float)
df_popu.info()
````

Unimos los dos dataframes, el de population y el del inicio(Que se me ólvido como se llama). 

````
df_merge = pd.merge(df_t_med[['Country','temperature','date']],
                    df_popu, on = ['Country','date'], how = 'inner')
````

````
df_merge
````

El merge anterior se agrego a los demás para que todo quede en una misma tabla. 


````
df_merge = pd.merge(df_merge,
                    df_co2e,
                    on = ['Country','name','date'],
                    how = 'inner')

df_merge = pd.merge(df_merge,
                    df_elec,
                    on = ['Country','name','date'],
                    how = 'inner')

df_merge = pd.merge(df_merge,
                    df_agri,
                    on = ['Country','name','date'],
                    how = 'inner')

df_merge = pd.merge(df_merge,
                    df_fore,
                    on = ['Country','name','date'],
                    how = 'inner')
df_merge
````
